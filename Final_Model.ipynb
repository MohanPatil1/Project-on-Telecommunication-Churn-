```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report,confusion_matrix
import pickle as pkl
```

```python
# Reading the dataset
df = pd.read_csv(r"C:\Users\Mohan Patil\OneDrive\Desktop\Datasets\Churn (1).csv",index_col='Unnamed: 0')
df
```

```python
df.drop(["area.code"],inplace=True,axis=1)
df.drop(["state"],inplace=True,axis=1)
df
```

```python
df.info()
```

```python
# Convert column to float beforing removing Nan value
column_convert = ['day.charge','eve.mins']
df[column_convert] = df[column_convert].astype(float)
```

```python
non_numeric_value = pd.to_numeric(df['day.charge'],errors='coerce').isnull().sum()
print(non_numeric_value)
```

```python
column_mean = pd.to_numeric(df['day.charge'],errors='coerce').mean()
df['day.charge'].fillna(column_mean,inplace=True) # Replacing the Nan Value with mean
```

```python
non_numeric_value1 = pd.to_numeric(df['eve.mins'],errors='coerce').isnull().sum()
print(non_numeric_value1)
```

```python
column_mean = pd.to_numeric(df['eve.mins'],errors='coerce').mean()
df['eve.mins'].fillna(column_mean,inplace=True) # Replacing the Nan Value with mean
```

```python
df.head(15)
```

```python
# To Convert Categorical class column in to numeric 
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['voice.plan']=le.fit_transform(df['voice.plan'])
df['intl.plan']= le.fit_transform(df['intl.plan'])
df['churn']=le.fit_transform(df['churn'])
```

```python
df.head()
```

```python
df.dtypes
```

```python
# Changing Column Name 
df1 = df.rename(columns={'account.length':'account_length','voice.plan':'voice_plan','voice.messages':'voice_messages','intl.plan':'international_plan',
                  'intl.mins':'international_mins','intl.calls':'international_calls','intl.charge':'international_charge',
                  'day.mins':'day_mins','day.calls':'day_calls','day.charge':'day_charge','eve.mins':'evening_mins','eve.calls':'evening_calls',
                  'eve.charge':'evening_charge','night.mins':'night_mins','night.calls':'night_calls','night.charge':'night_charge','customer.calls':'customer_calls'})
```

```python
df1.corr()
```

```python
# Outlier Treatment
# Considering columns whose outliers need to be treated
req_col = ['account_length','voice_messages','international_calls','day_mins', 'evening_mins', 'night_mins', 'international_mins', 'day_calls', 'day_charge', 'evening_calls', 
           'international_charge','evening_charge', 'night_calls', 'night_charge','customer_calls']
```

```python
# Calculate quartile And IQR 
q1 = np.percentile(df1[req_col],25)
q3 = np.percentile(df1[req_col],75)
iqr = q3 - q1

# Get Lower and upper bound
upper = q3+1.5*iqr 
lower = q1-1.5*iqr

outlier = ((df1[req_col]<lower) | (df1[req_col]>upper))

# Get the indices of outlier rows
outlier_indices = np.where(outlier)[0]

print("Total No. of Outlier:",len(outlier_indices))
print("\n","Outlier Indices :",outlier_indices)
```

```python
# Reset the index the dataframe
df_reset_index = df1.reset_index(drop=True)
# Remove row with outlier indices
df_clean = df_reset_index.drop(outlier_indices)
# print the modified dataframe
df_clean.info()
```

```python
# Feature Selection- Using Univariate Analysis
data = df_clean.drop(['churn'],axis=1)

from sklearn.feature_selection import SelectKBest,f_classif
select = SelectKBest(f_classif,k=9)

data_req = select.fit_transform(data,df_clean['churn'])
data_req
```

```python
df_new = pd.DataFrame(select.inverse_transform(data_req),index=df_clean.index,columns=data.columns)

req_columns = df_new.columns[df_new.var()!=0]
df_final = df_new[req_columns]

df_target = df_clean['churn']
Final_df = pd.concat([df_final, df_target], axis =1)

Final_df.describe()
```

# Balancing the Data

```python
# Oversampling Minor Class
from imblearn.over_sampling import SMOTE
```

```python
x = Final_df.iloc[:,:9]
y = Final_df['churn']
# Implementing Oversampling for Handling imbalanced
smk = SMOTE(sampling_strategy='minority',random_state=0)
x_res , y_res = smk.fit_resample(x ,y)
final_bal = pd.concat([pd.DataFrame(x_res),pd.DataFrame(y_res)],axis=1)
final_bal
```

```python
print(final_bal['churn'].value_counts())
```

## Build Model : Balanced Processed data post Feature Engineering 

```python
#Feature & Target Variables
X1 = final_bal.iloc[:, :9]
Y1 = final_bal['churn']

X1_train, X1_test, Y1_train, Y1_test = train_test_split(X1, Y1, test_size= 0.3, random_state=10)
```

```python
rf = RandomForestClassifier(random_state=0,class_weight='balanced')
rf.fit(X1_train,Y1_train)
# Train model 
y_pred_rf_train = rf.predict(X1_train)
acc_rf_train = accuracy_score(y_pred_rf_train,Y1_train)
print('Training Accuracy:',acc_rf_train*100)
# Test Model
y_pred_rf_test = rf.predict(X1_test)
acc_rf_test = accuracy_score(y_pred_rf_test,Y1_test)
print('Testing Accuracy:',acc_rf_test*100)
print('\n')
print(classification_report(y_pred_rf_test,Y1_test))
print('\n')
print(confusion_matrix(Y1_test,y_pred_rf_test))
```
